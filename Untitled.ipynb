```python
pi = 0.5
plt.figure()
y1 = np.random.normal(0, 1, size=500)
y2 = np.random.normal(0, np.exp(-3), size=500)
y = pi*y1 + (1-pi)*y2
sns.kdeplot(y, label="mix gauss", marker=".")
sns.kdeplot(y1, label="gauss1",)
sns.kdeplot(y2, label="gauss2",)
plt.ylim([0, 2])
plt.title("prior")
plt.savefig("figs/prior.png")
```

```python
def softplus(x):
    return np.log(1+np.exp(x))
def concat(x):
    target = []
    for arr in x:
        target.append(arr.flatten())
    return np.concatenate(target)
```

```python
plt.figure(figsize=[12, 6])
plt.subplot(2, 2, 1)
sns.kdeplot(softplus(concat(stat["rho"][0])), label="round0")
plt.title("std")
plt.xlim([0, 0.16])
plt.subplot(2, 2, 2)
sns.kdeplot(softplus(concat(stat["rho"][10])), label="round10")
plt.title("std")
plt.xlim([0, 0.16])
plt.subplot(2, 2, 3)
sns.kdeplot(softplus(concat(stat["rho"][50])), label="round50")
plt.title("std")
plt.xlim([0, 0.16])
plt.subplot(2, 2, 4)
sns.kdeplot(softplus(concat(stat["rho"][99])), label="round99")
plt.title("std")
plt.xlim([0, 0.16])
plt.savefig("figs/std_kde.png")
```
